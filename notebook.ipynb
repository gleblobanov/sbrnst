```python
import search
import matplotlib.pyplot as plt

from nltk.text import Text
```

```python
search_words = ['соборность', 'социабельность', 'социальность', 'общежительность', 'всеединство']
```

```python
# Гурвич
```

```python
corpus = search.create_corpus(1)
```

```python
text = Text(corpus.words())
```

```python
corpus.fileids()
```

```python
search_words = [w for w in text if w.startswith('собор')]
plt.figure(figsize=(10, 20))
text.dispersion_plot(search_words)
print(search_words)
```

```python
search_words = [w for w in text if w.startswith('социаб')]
plt.figure(figsize=(10, 40))
text.dispersion_plot(search_words)
print(search_words)
```

```python
search_words = [w for w in text if w.startswith('социальн')]
plt.figure(figsize=(10, 150))
text.dispersion_plot(search_words)
print(search_words)
```

```python
search_words = [w for w in text if w.startswith('общежит')]
plt.figure(figsize=(10, 10))
text.dispersion_plot(search_words)
print(search_words)
```

```python
search_words = [w for w in text if w.startswith('всеедин')]
plt.figure(figsize=(10, 20))
text.dispersion_plot(search_words)
print(search_words)
```
